```python
train_target = data['1'].to_numpy()
```

```python
train_target
```

```python
from sklearn.model_selection import train_test_split

data, val_data, target, val_target = train_test_split(train_data,
                                                     train_target,test_size=0.1,
                                                     random_state=42)
```

```python
data.shape
```

```python
model = keras.Sequential()
# 첫 번째 
model.add(keras.layers.Conv2D(8, kernel_size=3, strides=1,
                             padding='same', input_shape=(128,128,3)))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(keras.layers.MaxPool2D(2, strides=2))

# 두 번째
model.add(keras.layers.Conv2D(16, kernel_size=3, strides=1,
                             padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(keras.layers.MaxPool2D(2, strides=2))

# 세 번째
model.add(keras.layers.Conv2D(32, kernel_size=3, strides=1,
                             padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(keras.layers.MaxPool2D(2, strides=2))

# 4 번째
model.add(keras.layers.Conv2D(64, kernel_size=4, strides=1,
                             padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(keras.layers.MaxPool2D(2, strides=2))

# fully connected 
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(100))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(keras.layers.Dropout(0.4))
model.add(keras.layers.Dense(1))
```

```python
model.summary()
```

```python
# adam = optimizers.Adam(lr = 0.001)
adam = tf.keras.optimizers.Adam(learning_rate=0.001)

model.compile(optimizer=adam,loss='mean_absolute_error',
              metrics='accuracy')
checkpoint_cb = keras.callbacks.ModelCheckpoint('best-cnn-model.h5',
                                                save_best_only=True)
early_stopping_cb = keras.callbacks.EarlyStopping(patience=2,
                                                  restore_best_weights=True)
history =  model.fit(data, target, batch_size=8, epochs=20,
                     validation_data=(val_data, val_target),
                     callbacks=[checkpoint_cb,early_stopping_cb])
```
